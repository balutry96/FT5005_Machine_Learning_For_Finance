```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats.mstats import winsorize

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# 1. Create a simulated dataset
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1, n=1000

print(type(data))

# 2. Draw a boxplot and a histogram of that variable
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data)
plt.title('Boxplot of Normally Distributed Variable')

plt.subplot(1, 2, 2)
sns.histplot(data, kde=True)
plt.title('Histogram of Normally Distributed Variable')

plt.tight_layout()
plt.show()

# 3. Winsorize the variable at both ends by 10%
winsorized_data = winsorize(data, limits=[0.1, 0.1])  # 10% at both ends

# 4. Draw the boxplot after winsorization
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(winsorized_data)
plt.title('Boxplot of Normally Distributed Variable')

plt.subplot(1, 2, 2)
sns.histplot(winsorized_data, kde=True)
plt.title('Histogram of Normally Distributed Variable')

plt.tight_layout()
plt.show()
```
